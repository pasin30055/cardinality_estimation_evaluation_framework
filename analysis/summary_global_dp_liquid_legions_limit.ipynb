# Input parameters

Users must specify where the raw results are by the `in_dir`, and where to store the output results by the `out_dir` which could be the same as the `in_dir`.

````
in_dir = 'om/repo/ll_limit/eval_results' #@param
out_dir = 'om/repo/ll_limit/eval_results' #@param
````

````
# We plot several curves, and each of which is corresponding to a criterion.
# For example, [.95, .05] means 5% relative error within 95% of the runs.
# Users can customize the criteria list if they want.
ESTIMABLE_CRITERIA = [[.95, .05], [.90, .10], [.80, .20]] #@param
ESTIMABLE_CRITERIA
````

````
import itertools
import os

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

import wfa_cardinality_estimation_evaluation_framework.evaluations.analyzer as analyzer

plt.rcParams['figure.figsize'] = [12, 8]
````

````
a = analyzer.CardinalityEstimatorEvaluationAnalyzer(
    out_dir=out_dir,
    evaluation_directory=in_dir,
    evaluation_run_name='results',
    evaluation_name='global_dp_stress_test',
    estimable_criteria_list=ESTIMABLE_CRITERIA,
)
````

````
def parse_row(row):
  reach = int(row['scenario'].split('-')[1].lstrip('reach:'))
  sketch_estimator_name = row['sketch_estimator'].split('-')

  global_dp_str = sketch_estimator_name[4]
  if global_dp_str == 'no_global_dp':
    epsilon = 'inf'
  else:
    epsilon = global_dp_str.split('_')[2]

  sketch_config = sketch_estimator_name[1]

  results = pd.Series({
      'sketch_config': sketch_config,
      'run_index': row['run_index'],
      'relative_error': row['relative_error_1'],
      'reach': reach,
      'epsilon': epsilon,
  })
  return results

df = a.raw_df.apply(parse_row, axis=1)
df
````

````
def agg_estimable(df):
  relative_errors = df['relative_error']
  re_bound = set([x[1] for x in ESTIMABLE_CRITERIA])
  results = {
      'relative_error_mean': np.mean(relative_errors),
      'relative_error_std': np.std(relative_errors),
  }
  for re in re_bound:
    results.update({
        f'estimable_prop_under_re_{re}': np.mean(np.abs(relative_errors) < re)
    })
  return pd.Series(results)


df_agg = (
    df.groupby(['sketch_config', 'reach', 'epsilon'])
    .apply(agg_estimable)
)
df_agg.head(20)
````

````
summary_df_list = []
for criterion in ESTIMABLE_CRITERIA:
  prop_runs, re_bound = criterion[0], criterion[1]
  re_col = f'estimable_prop_under_re_{re_bound}'
  summary_df = (
      df_agg.loc[df_agg[re_col] >= prop_runs].reset_index()
      .groupby(['sketch_config', 'reach']).agg({'epsilon': 'min'}).reset_index()
  )
  summary_df['reach'] = summary_df['reach'].astype('float')
  summary_df['epsilon'] = summary_df['epsilon'].astype('float')
  summary_df['criterion'] = f'{prop_runs}|{re_bound}'
  summary_df_list.append(summary_df)

summary_df = pd.concat(summary_df_list, ignore_index=True)
summary_df
````

````
summary_df.to_csv(os.path.join(out_dir, 'summary_df.csv'), index=False)
````

````
for conf in summary_df['sketch_config'].unique():
  fig, ax = plt.subplots()
  sns.lineplot(x='reach', y='epsilon', hue='criterion',
               data=summary_df.loc[summary_df['sketch_config'] == conf],
               ax=ax)
  ax.set_title(f'Liquid Legions parameters: {conf}')
  ax.set_xscale('log')
  ax.set_yscale('log')
  ax.set_xlabel('Audience size')
  ax.set_ylabel('Epsilon')
````
